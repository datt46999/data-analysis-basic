# import library

```python
import numpy as np
import pandas as pd
import seaborn
import matplotlib.pyplot as plt

```

## Cau 1

```python
# laod data
data = pd.read_csv("data\\titanic_disaster.csv")
# show 10 dong dau
data.head(10)
```

## Cau 2

```python
# kiem tra gia tri rong = isna tra ve true or flue
Missvalu=data.isna()
Missvalu
```

```python
# truc quan du lieu bang bieu do heatmap
seaborn.heatmap(Missvalu)
```

## Cau 3

```python
# tao ra hai cot moi la firstName va secondNAme tu co Name 
# /dung ham str.split de phan tac chuoi trong cot Name tanh 2 cot
# dung .drop de xoa cot name
data[["firstName","secondName"]]=data["Name"].str.split(",",expand=True)
data=data.drop("Name", axis=1)
data.head(10)
```

## Cau 4

```python
# tao 1 bien lieu cac ca gia tri trong cot Sex cua tap data
#  diet cac phan tu trong bien vua tao neu la male su lai thanh M con lain se la F
# thay th cot Sex bang bien vua duoc sua
Sex = data["Sex"]
for i in range(0,len(Sex)):
    if Sex[i]=="male":
        Sex[i]="M"
    else: Sex[i]="F"
data["Sex"]=Sex
data.head(10)
```

```python

seaborn.boxplot(x="Age",y="Embarked", data=data)
```

## cau 5

```python
# tao 1 bien dee tinh gia tri trung binh cua cot Age lang ham mean()
# fill tat ca cac gia tri null bang gia tri TB da tinh truoc do, show 10 dong dau
mean_age=int(data["Age"].mean()) 
data["Age"].fillna(value=int(mean_age), inplace=True)
data.head(10)
```

```python
# seaborn.heatmap(data=mea,)
```

## cau 6

```python
# tao 1 bien luu cac gia tri cua cot Age trong data
# dieu cac bien neei <= 12 == Kid
# (12,18] Teen
# (18,60] Adult
# con lai la Older 
ageGroup=data["Age"]
tmp=[]
for i in range(len(ageGroup)):
    if ageGroup[i]<=12: tmp.append("Kid")
    elif  ageGroup[i]<=18: tmp.append("Teen")
    elif ageGroup[i]<=60: tmp.append("Adult")
    else: tmp.append("Older")
data["Agegroup"]=tmp
data.head(10)
```

## cau 7


```python
# dung ham str.split(".") de tac Mr, Mrs, Miss ra khoi ten
# tao 1 cot moi de luu cac gia trị Mr, Mrs, Miss
# show 10 dong dau

df_tmp = data["secondName"].str.split(".",expand=True)
data["namePrefix"] = df_tmp[0]
data.head(10)
```

## cau 8

```python
# tao 1 list rong 
#  tao 2 bieen tuong ung vs hai cot SibSp va Parch
# List rong sex chua cac gai tri theo cong thuc 1+sib[i]+par[i] 
# tao cot familySize trong data tuong ung de luu cac gia tri trong bien rong
# shiw 10 dong dau
tmp=[]
sib=data["SibSp"]
par=data["Parch"]
for i in range(len(data["SibSp"])):
    cur = 1 +sib[i]+par[i]
    tmp.append(cur)
data["familySize"]=tmp
data.head(10)
```

## cau 9

```python
# tao 1 list rong
# tao bien chua cac gia tri cua cot familySize
# duyet qua cac bien do , them vao list da tao o tren neu gia tri i tai cua cot familySize ==0 nguoc lai ta them 1
# show 20 dong dau
tmp = []
Alone= data["familySize"]
for i  in range (len(Alone)):
    tmp.append(1) if Alone[i]==0 else tmp.append(0)
data["Alone"]=tmp
data.head(20)
```

## cau 10

```python
# tao 1 list rong
# tao bien chua cac gia tri cua cot Cabin
# appeng vao list rong <Unknown> neu gia tri la rong nguoc laij tra ve kieu gai tri thu i cu cot Cabin
# show 20 dong dau
tmp = []
CB= data["Cabin"]
for i  in  CB.values:
    tmp.append("Unknown") if i is np.nan else tmp.append(i[0])
    #tmp.append('Unknown' if i is np.nan else i[0])
data["typeCabin"]=tmp
data.head(20)
```
