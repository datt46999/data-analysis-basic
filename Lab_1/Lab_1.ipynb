# import library

```python
import numpy as np
import pandas as pd
```

Show First and lasr 10 lines 

```python
df = pd.read_csv("dulieuxettuyendaihoc.csv")
# show first 10 line
# df.head(10)

# show last 10 line 
df.tail(10)
```

fill missing value in column DT with value 0

```python
df['DT'] = df['DT'].fillna(0)
df.tail(10)
```

Using mean method to replace missing value of  column T1

```python
mean_value = df['T1'].mean()
# replace original data ; implace = True
df['T1'].fillna(value=mean_value, inplace=False) 
# print(df)
```

Handle missing value by mean method in all data

```python
is_flaot64 = df.dtypes==float
df_column_list = df.columns[is_flaot64].tolist()
df_column_list
for col in df_column_list:
    df[col].fillna(df[col].mean(), inplace=True)
df.head(10)
```

average columns with formula:
 TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10 

```python
for i in ["1","2","3"]:
    df["TBM"+i] = (df['T'+i]*2 + df['L'+i] + df['H'+i] + df['S'+i] + df['V'+i]*2 + df['X'+i] + df['D'+i] + df['N'+i])/10
# df.rename(columns={"TBM6":"TBM3"}, inplace=False)
df.head(10)
```

classify TBM 

```python
for i in ["1", "2", "3"]:
    df["TBM" + i] = df["TBM" + i].astype(float)

for i in ["1", "2", "3"]:
    df["XL" + i] = np.where(df["TBM" + i] < 5.0, "Y",
                  np.where(df["TBM" + i] < 6.5, "TB",
                  np.where(df["TBM" + i] < 8.0, "K",
                  np.where(df["TBM" + i] < 9.0, "G", "XS"))))
df.head(20)
```

transfer average score from 10. to 4.

```python
def min_max_nomal(colVa):
    return (colVa/10)*4

for i in ["1", "2", "3"]:
    df["TBM" + i] = df["TBM" + i].astype(float)
    
for i in ["1", "2", "3"]:
    df["US_TBM" + i] = min_max_nomal(df["TBM" + i])
df.head(20)
```
