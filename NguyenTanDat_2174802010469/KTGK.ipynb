```python
# cau 1
# create data 
data={"Name" : ["Alice", "Bob", "Chirlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ivan", "Jack", "Kelly", "Liam", "Mona", "Nina", "Osca"],
        "Age" :[25, 30, 35,28 ,22 ,45 ,34 ,31, 27, 29, 33, 40, 26, 32, 36 ] ,
        "Salary": [50000, 60000, 70000, 55000, 52000, 80000, 72000, 68000, 61000, 59000, 63000, 77000, 53000, 66000, 75000] 
}

df = pd.DataFrame(data)

df.head(15)
```

```python
# cau 11
# kiem tran dong dataframe co gia trin NaN hay ko
check_NaN=  df.isnull().values.any()
check_NaN
```

```python
# cau 12
# thay the gia tri co Age > 30 bang cach tinh gia tri trung binh cua cot do
mean_Age = df["Age"].mean()
age= df["Age"]
for i in range(0,len(age)):
    if age[i]>30:
        age[i] = mean_Age
df["Age"]=age
df
```

```python
# cau 13
# normalize cot Age tiu khoang 0 toi 1


normalize= df["Age"]
for i in range(0,len(normalize)):
    if (normalize[i]==mean_Age): normalize[i] = 1.0
    else: normalize[i] = normalize[i] / mean_Age 
normalize
```

```python
# cau 14
# phan loai thanh cac nhom Young, MIddle-aged, Old
tmp = []
age = df["Age"]
for i in age.values:
    if i< 30: tmp.append("Young")
    elif i >= 30 and i<60: tmp.append("Middled-aged")
    else: tmp.append("Old")
df["Age_Group"] = tmp
df
```

```python
# cau 15
# ti le phan tram thay doi cua cot Salary
per_chance= df["Salary"].pct_change()
per_chance
```

```python
# cau 16
# tim tat ca cac gia ti trung lap tren cot Name xoa vong lap giu hang dau tien
df.drop_duplicates(subset='Name',keep="first")
df
```
