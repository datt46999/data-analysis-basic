# Import library

```python
import numpy as np 
import pandas as pd
```

## Read data

```python
# doc du lieu
data = pd.read_csv("patient_heart_rate.csv",encoding='latin-1')
# doc 10 hang dau cua du lieu
data.head(10)
```

## Vấn đề 1: Tiến hành tải dữ liệu vào chương trình ứng dụng Python và giải quyết vấn đề ―Missing header in the csv file‖


```python
# the hearder vao dataframe
colmn_name = ["Id","Name","Age","Weight","m0006","m0621","m1218","f0006","f0612","f1218"]

data= pd.read_csv("patient_heart_rate.csv",encoding="latin-1",names=colmn_name)
# # hien thi 10 hang dau
data.head(10)
```

##  Van de 2 :  Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột ―Name‖ chứa bao gồm ―Firstname‖ và ―Lastname‖, giải pháp là ta sẽ tách ra làm 2 cột

```python
# tac cot Name thanh hai cot Fristname va Lastname
data[["Firstname","Lastname"]] = data["Name"].str.split(expand=True)
# xoa cot name
data= data.drop("Name", axis=1)
# hine thi 10 hang dau
data.head(10)
```

##   Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn ―kg‖

```python
# lay gia tri cot weight
weight = data["Weight"]
for i in range(0, len(weight)):
    # lay gi tri can nang
    x=str(weight[i])
    if "lbs" in x[-3:]:
        # xoa 3 phan tu cuoi
        x=x[:-3:]
        float_x=float(x)
        # doi don vi
        y=int(float(float_x/2.2))
        y=str(y)+"kgs"
        weight[i]=y
data.head(10)
```

## Vấn đề 4: Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp có thể đưa ra là xóa bỏ

```python
# xoa cac dong du lieu co gia tri la NaN
data.dropna(how="all", inplace=True)
#  hien thiu 10 hang dau
data.head(10)
```

## Vấn đề 5: Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn[fullname, lastname, age, weight,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phải dựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý.

```python
# xu lu du lieu bi trung lap
data = data.drop_duplicates(subset=["Firstname","Lastname", "Age", "Weight"])
# hien tji 10 dong dau
data.head(10)
```

## Vấn đề 6: Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII. 
Giải pháp: 

```python
# ching sua loi non-ASCI
data.Firstname.replace({r"[^\x00-\x7F]+":""}, regex=True, inplace=True)
data.Lastname.replace({r"[^\x00-\x7F]+":""}, regex=True, inplace=True)
#  hien thi 10 dong dau
data.head(10)
```

## Vấn đề 7: ―Missing values‖, vấn đề này xảy ra tại các cột ―Age‖, ―Weight‖ và ―Heart Rate‖. 

```python
# thong ke du lieu thi owr cot Age va cot Weight
index_null = data[data["Age"].isna() & data["Weight"].isna()].index
index_null
```

```python
# gia tri age trong cot khi thieu thay the bang gia tri PP mean
data["Age"].fillna(data["Age"].mean(), inplace=True)
# hien thui 10 donmg dau
data.head(10)
```

## Vấn đề 8: ―một cột chứa quá nhiều thông tin cần được phân rã‖, như trong bài toán này ta thấy header ―m0006‖ chứa các nội dung bao gồm: m  male, 1218 ~ 12-18 (mm-dd). Còn giá trị thì là kết quả huyết áp.


```python

# tach ca cot phan ra tanh ca cot PulseRate, sex_and_time
data= pd.melt(data,id_vars=["Id","Age","Weight","Firstname","Lastname"], value_name='PulseRate', var_name="sex_and_time").sort_values(["Id", "Age","Weight","Firstname", "Lastname"])
data.head(10)
```

```python
# extract sexm hour lower bound, and hours upper bound group
tmp_df= data["sex_and_time"].str.extract("(\D)(\d+)(\d{2})", expand=True)
tmp_df.head(10)
```

```python
# gan ten cot 
tmp_df.columns =["Sex", "hour","minute"]
tmp_df.head(10)
```

```python
# tao cot tiem du tren cot hour va minute
tmp_df["Time"] = tmp_df["hour"]+"-"+tmp_df["minute"]
tmp_df.head(10)
```

```python
# gop vao df
data= pd.concat([data,tmp_df],axis=1)
data.head(10)
```

```python
data = data.drop(["sex_and_time","hour","minute"],axis=1)
data = data.dropna()
data.head(10)
```

```python
data.to_csv("outputcleanup.csv", index=False)
```

```python
def avgPulserate(df, firstname):
  sumPulserate = 0
  nsumPulserate = 0
  for i, value in enumerate(data['Firstname']):
    if value == firstname:
      try:
        if data['PulseRate'][i] !='0':
          sumPulserate += int(data['PulseRate'].values[i])
          nsumPulserate += 1
      except:
        pass
  if nsumPulserate > 0:
    return str(sumPulserate/nsumPulserate)
  else:
    return 0
```

```python
def isValidIndex1(index, nrows):
  return index - 1 > 0 & index + 1 < nrows

def isValidIndex2(index, nrows):
  return index + 1 < nrows & index + 2 < nrows

def isPulseRateNotNull1(index, df):
  return str(df['PulseRate'].values[index-1])!= '0' and str(df['PulseRate'].values[index+1])!= '0'

def isPulseRateNotNull2(index, df):
  return str(df['PulseRate'].values[index+1])!= '0' and str(df['PulseRate'].values[index +2])!= '0'


def isInGroup1(df, index, firstname):
  return df['Firstname'].values[index-1]== firstname and df['Firstname'].values[index +1]== firstname

def isInGroup2(df, index, firstname):
  return df['Firstname'].values[index+1]== firstname and df['Firstname'].values[index +2]== firstname

def updatePulseRatena (index, df):
  nrows = len(df['PulseRate']) -2
  newvalue = ''
  firstname = df['Firstname'].values[index]
  try:
    if isValidIndex1(index,nrows) & isPulseRateNotNull1(index, df) & isInGroup1(df, index, firstname):
      try:
        newvalue = (int(df['PulseRate'].values[index-1]) + int(df['PulseRate'].values[index+1]))/2
      except:
        newvalue = '0'
    elif isValidIndex2(index,nrows) & isPulseRateNotNull2(index, df) & isInGroup2(df, index, firstname):
      try:
        newvalue = (int(df['PulseRate'].values[index-1]) + int(df['PulseRate'].values[index+1]))/2
      except:
        newvalue ='0'
    else:
      newvalue = avgPulserate(df, firstname)
  except:
      newvalue = avgPulserate(df, firstname)
  finally:
    pass
  return newvalue
```

```python
pulserateupdate = []
for index, value in enumerate(data["PulseRate"]):
  if value =='0' or value == '':
    newvalue = updatePulseRatena(index,data)
    pulserateupdate.append(newvalue)
  else:
    pulserateupdate.append(value)
data["PulseRateUpdate"] = pulserateupdate
data
```
